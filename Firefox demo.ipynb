## Firefox Demo

Demo of using headless Firefox with selenium webdriver in a MyBinder run container.

## Setup
Dependencies are:

- desktop applications
  - Firefox (eg `apt-get install -y firefox`)
- Python packages
  - `webdriverdownloader`
  - `selenium`
  - *for example, `pip3 install webdriverdownloader selenium`*
  
The webdriver can be downloaded and installed via the Python `webdriverdownloader` package:

```python
#!/usr/bin/env python3

from webdriverdownloader import GeckoDriverDownloader
gdd = GeckoDriverDownloader()
gdd.download_and_install("v0.23.0")
```

Setup is managed using files in the `binder/` directory.

## Example of Using Headless Firefox to Grab a Screenshot from a Downloaded Page

We can grab a web page for scraping, or taking screenshots, using `selenium` and the downloaded websdriver.

To get the driver instance, we need to open a Firefox webdriver with the `headless` option set as `True`.

````python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options

options = Options()
options.headless = True

#If we need to set the path to geck, we can...
#browser = webdriver.Firefox(executable_path=geckobin, options=options)

browser = webdriver.Firefox(options=options)
````

## Grab a Page

Define the page URL, optionally set the browser size, and http get the URL.

````python
url = 'https://www.bbc.com'
````

````python
browser.set_window_size(800, 400)
browser.get(url)
````

## Take a Screenshot of a Downloaded Page

We can save a screenshot of the current borwser window to a file and then display that image file in the notebook:

````python
outfile = 'test.png'
````

````python
browser.save_screenshot( outfile )
# Close the browser instance
browser.close()
````
